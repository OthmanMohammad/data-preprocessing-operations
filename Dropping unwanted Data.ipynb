Python | Delete rows/columns from DataFrame using Pandas.drop()

```python
import pandas as pd
```

Example #1: Dropping Rows by index label

```python
df = pd.read_csv('C:/Users/ASUS/Desktop/nba.csv',index_col ="Name" ) 

# dropping passed values 
df.drop(["Avery Bradley", "John Holland", "R.J. Hunter"], inplace = True) 
  
# display 
df.head()
```

Example #2 : Dropping columns with column name

```python
df = pd.read_csv('C:/Users/ASUS/Desktop/nba.csv',index_col ="Name" ) 

  
# dropping passed columns 
df.drop(["Team", "Weight"], axis = 1, inplace = True) 
  
# display 
df 
```
